```python
import numpy as np
import seaborn as sns
import pandas as pd

#annoying issue with changes to matplotlib, this gives the lines to separate
import matplotlib as mpl
mpl.rcParams['patch.force_edgecolor'] = True

%matplotlib inline
```

```python
df1 = pd.read_csv('df1', index_col=0)
```

```python
df1.head()
```

```python
df2 = pd.read_csv('df2')
```

```python
df2.head()
```

```python
df1['A'].hist(bins=30)
```

```python
# another way to make a histogram
df1['A'].plot(kind='hist', bins=30)
```

```python
# Even shorter method is through this
df1['A'].plot.hist(bins=30)
```

```python
df2.head()
```

```python
df2.plot.area(alpha=0.4)
```

```python
# he recommends using seaborn for these things
df2.plot.bar()
```

```python
df2.plot.bar(stacked=True)
```

```python
df1.head()
```

```python
# this is something if you want to look at stocks over time
# You can apply a lot of matplotlib elements here
df1.plot.line(x=df1.index, y='B', figsize=(12,3), lw=1)
```

```python
# We are showing 3 levels of information
df1.plot.scatter(x='A', y='B', c='C', cmap='coolwarm')
```

```python
# if we didnt want to compare by color and by size instead
df1.plot.scatter(x='A', y='B', s=df1['C']*100, edgecolor='black')
```

```python
# simple pandas boxplot
df2.plot.box()
```

```python
df = pd.DataFrame(np.random.randn(1000,2), columns=['a','b'])
```

```python
# basically scatterplot but with hexagons
df.plot.hexbin(x='a',y='b', gridsize=25, cmap='coolwarm')
```

```python
# df2['a'].plot.density() # is the exact same thing
df2['a'].plot.kde()
```

```python
df2.plot.density()
```
